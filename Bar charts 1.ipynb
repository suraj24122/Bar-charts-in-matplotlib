```python
import matplotlib.pyplot as plt
```

```python
# simple verticle bar chart
subjects = ['Math', 'Science', 'English', 'History', 'Computer']
marks = [88, 92, 80, 76, 95]
```

```python
plt.bar(subjects, marks, color = "red")
plt.title("Marks by subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
# verticle bar chart with value labels
for i in range(len(subjects)):
    plt.text(i, marks[i] + 1, str(marks[i]), ha="center", va="bottom")
```

```python
# setting bar width
import numpy as np
cities = ['Delhi', 'Mumbai', 'Chennai', 'Kolkata']
pollution = [85, 90, 75, 80]

x = np.arange(len(pollution))
width = 0.5

plt.bar(x, pollution, width = width, label = "pollution")
plt.xlabel("cities")
plt.ylabel("pollution")
plt.title("pollution per cities")
plt.xticks(x, cities)
plt.legend()
plt.tight_layout()
```

```python
# side by side bar charts
import numpy as np
years = ['2020', '2021', '2022']
product_A = [100, 120, 130]
product_B = [90, 110, 125]
# getting position of the each year
x = np.arange(len(years))
width = 0.35
# plot side by side bars
plt.bar(x - width/2, product_A, width, label = "Product A", color = "blue")
plt.bar(x + width/2, product_B, width, label = "Product B", color = "green")
plt.xlabel("years")
plt.ylabel("Sales")
plt.title("Products and sales comparision")
plt.xticks(x, years)
for i in range(len(years)):
    plt.text(x[i] - width/2, product_A[i] + 1, str(product_A[i]), ha = "center")
    plt.text(x[i] + width/2, product_B[i] + 1, str(product_B[i]), ha = "center")

plt.legend()
plt.tight_layout()
```

```python
# horizontal bar charts using plt.barh()
languages = ['Python', 'Java', 'C++', 'JavaScript', 'Go']
popularity = [95, 88, 75, 85, 65]

plt.barh(languages, popularity, color = "red")
plt.title("Programming language popularity")
plt.grid(True)
plt.xlabel("Popularity (%)")
for i in range(len(languages)):
    plt.text(popularity[i] + 1, i, str(popularity[i]), va = "center")
plt.tight_layout()
plt.show()
```
